```python
import numpy as np
import matplotlib.pyplot as plt
from backend.processclass import Experiment2D
from backend.experimentsclass import loop_param
from backend.plotting import plot_from_exps
```

```python
pr = Experiment2D()
# Initializing model
pr.n0 = 2.7  # 1/nm^2
pr.F = 100.0  # 1/nm^2/s
pr.s = 1.0
pr.V = 0.05  # nm^3
pr.tau = 700e-6  # s
pr.D = 8e5  # nm^2/s
pr.sigma = 0.005 # nm^2
pr.f0 = 1e6
pr.fwhm = 20  # nm
pr.order = 1
pr.step = 0.1  # nm
```

```python
vals = np.arange(0, 8, 1)
vals = np.power(10, vals)
param_name = 'D'
exps = loop_param(param_name, vals, pr)
exps.param = param_name
r_max = exps.r_max
twin_peak = exps.get_peak_twinning()
exps.plot()
```
