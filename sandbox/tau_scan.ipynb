```python
import numpy as np
import matplotlib.pyplot as plt
from backend.processclass import Experiment1D
from backend.experimentsclass import loop_param
from backend.plotting import plot_from_exps as plot
```

```python
pr = Experiment1D()
# Initializing model
pr.n0 = 2.7  # 1/nm^2
pr.F = 230.0  # 1/nm^2/s
pr.s = 0.01
pr.V = 0.01  # nm^3
pr.tau = 200e-6  # s
pr.D = 8e5  # nm^2/s
pr.sigma = 0.02  # nm^2
pr.f0 = 1.0e7
pr.fwhm = 50  # nm
pr.order = 1
pr.step = 0.5  # nm
```

```python
vals = np.arange(50e-6, 500e-6, 20e-6).round(6)
param_name = 'tau'
exps = loop_param(param_name, vals, pr)
exps.param = param_name
r_max = exps.r_max
twin_peak = exps.get_peak_twinning()
exps.plot()
```
