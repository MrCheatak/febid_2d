Comparison between simulated reference and found solution

```python
import numpy as np
import matplotlib.pyplot as plt
from processclass import Experiment2D
```

Problem 1

```python
pr = Experiment2D()
# Initializing model
pr.n0 = 2.7  # 1/nm^2
pr.F = 730.0  # 1/nm^2/s
pr.s = 1.0
pr.V = 0.05  # nm^3
pr.tau = 200e-6  # s
pr.D = 12e5  # nm^2/s
pr.sigma = 0.02  # nm^2
pr.f0 = 1.0e7
pr.fwhm = 50  # nm
pr.order = 1
pr.step = 0.5  # nm
```

```python
pr.solve_steady_state()
R = pr.R
r = pr.r
fig,ax = plt.subplots()
ax.plot(r, R, label='Reference')

pr.D=1.92884148e+06
pr.sigma=3.24744820e-02
pr.tau = 1.21525179e-04
pr.solve_steady_state()
R = pr.R
r = pr.r
ax.plot(r, R, label='Solution')
plt.legend()
plt.show()
```
