```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname='tau_p.txt'
data1 = np.genfromtxt(fname, delimiter='\t', skip_header=1)
fix, ax = plt.subplots()
x, y, color = data1[:,0], data1[:, 1], data1[:,2]
line = ax.scatter(x, y, c=color, s=1 , cmap='magma',)
ax.set_xlabel(r'$p_{out}$')
ax.set_ylabel(r'$tau_r$')
plt.title('Peak position map')
```

```python
fname='tau_p_1.txt'
data2 = np.genfromtxt(fname, delimiter='\t', skip_header=1)
fix, ax = plt.subplots()
x, y, color = data2[:,0], data2[:, 1], data2[:,2]
line = ax.scatter(x, y, c=color, s=2, cmap='magma',)
ax.set_xlabel(r'$p_{out}$')
ax.set_ylabel(r'$tau_r$')
plt.title('Peak position map')
plt.show()
```

```python
fname='tau_p_2.txt'
data3 = np.genfromtxt(fname, delimiter='\t', skip_header=1)
fix, ax = plt.subplots()
x, y, color = data3[:,0], data3[:, 1], data3[:,2]
line = ax.scatter(x, y, c=color, s=2, cmap='magma',)
ax.set_xlabel(r'$p_{out}$')
ax.set_ylabel(r'$tau_r$')
plt.title('Peak position map')
plt.colorbar(line, ax=ax)
plt.show()
```

```python
data = np.vstack((data1, data2))
fig, ax = plt.subplots()
ind = np.logical_and(data[:,1]<200,data[:,0]<4)
data = data[ind, :]
x, y, color = data[:,0], data[:, 1], data[:,2]
line = ax.scatter(x, y, c=color, s=2, cmap='magma')
ax.set_xlabel(r'$p_{out}$')
ax.set_ylabel(r'$tau_r$')
plt.title('Peak position map')
plt.colorbar(line, ax=ax)
plt.show()
```
