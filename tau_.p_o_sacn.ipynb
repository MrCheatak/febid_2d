```python
pr = Experiment2D()
# Initializing model
pr.n0 = 2.7  # 1/nm^2
pr.F = 730.0  # 1/nm^2/s
pr.s = 1.0
pr.V = 0.05  # nm^3
pr.tau = 200e-6  # s
pr.D = 8e5  # nm^2/s
pr.sigma = 0.02  # nm^2
pr.f0 = 1.0e7
pr.fwhm = 50  # nm
pr.order = 1
pr.step = 0.5  # nm
```

```python
def write_to_file(p_o, tau_r, r_max):
    import os
    fname = 'tau_p.txt'
    if not os.path.exists(fname):
        with open(fname, mode='x') as f:
            f.write('tau_r\tp_o\tr_max\n')
    else:
        with open(fname, mode='a') as f:
            for i in range(p_o.size):
                f.write(f'{p_o[i]}\t{tau_r[i]}\t{r_max[i]}\n')
param_name1 = 'D'
param_name2 = 'f0'
vals1 = np.arange(2e7, 3e8, 5e6)
vals2 = np.arange(5, 8.5, 0.1)
vals2 = np.power(10, vals2).astype(int)
data = np.zeros((vals1.shape[0]*vals2.shape[0], 3))
exps_all = []
temp = pr.__getattribute__(param_name1)
for i, val in enumerate(vals1):
    pr.__setattr__(param_name1, val)
    exps = loop_param(param_name2, vals2, pr)
    exps_all.append(exps)
    r_max, twin_peak = get_r_maxes(exps)
    tau_r = np.asarray([pr.tau_r for pr in exps])
    p_o = np.asarray([pr.p_o for pr in exps])
    write_to_file(p_o, tau_r, r_max)
pr.__setattr__(param_name1, temp)
```
